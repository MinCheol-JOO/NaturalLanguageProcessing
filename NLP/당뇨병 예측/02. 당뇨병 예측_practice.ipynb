# 데이터셋 분할


* 훈련 데이터셋(training set): 신경망을 훈련시키는 데 사용
* 검증 데이터셋(validation set): hyperparameter(초매개변수)를 튜닝하는 과정으로 사요
* 테스트 데이터셋(test set): 신경망의 성능을 최종 평가하는 데 사용
![](https://blog.kakaocdn.net/dn/R5F4R/btqHb0lGapZ/Cupp4u4JJLbKXKKhfgfhM1/img.png)
*https://ysyblog.tistory.com/69*

```python
# scikit-learn train_test_split으로 training, validation, test set 만들기
from sklearn.model_selection import train_test_split

X = dfs.loc[:, dfs.columns != '당뇨']
y = dfs.loc[:, '당뇨']

# train:test - 8:2 데이터 분할
# test:val - 8:2 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

# 다중 레이어 퍼셉트론

![img](https://i.imgur.com/ou4obJs.png)

## 케라스로 신경망 모델 만들기


```python
from tensorflow.keras.models import Sequential
model = Sequential()
```

```python
from tensorflow.keras.layers import Dense

# 히든 레이어1
model.add(Dense(units=8, activation='relu', input_dim=8))
# 히든 레이어2
model.add(Dense(units=8, activation='relu'))
# 출력 레이어
model.add(Dense(units=1, activation='sigmoid'))
```

## 모델 컴파일
* 손실 함수 : 훈련 데이터에서 신경망의 성능을 측정하는 방법으로 네트워크가 옳은 방향으로 학습될 수 있도록 도와 줍니다.
* 옵티마이저: 입력된 데이터와 손실 함수를 기반으로 네트워크를 업데이트하는 메커니즘입니다.
* 훈련과 테스트 과정을 모니터링할 지표 : 여기에서는 정확도(정확히 분류된 이미지의 비율)만 고려하겠습니다.

```python
# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## 모델 훈련

* 케라스에서는 fit 메서드를 호출하여 훈련 데이터에 모델을 학습시킵니다:

```python
# 학습하기
history = model.fit(X_train, y_train, epochs=1, verbose=False, validation_data=(X_val, y_val))
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(211)
plt.title('Loss')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()

plt.subplot(212)
plt.title('Accuracy')
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.legend()
```

## 결과 분석

## 테스트 정확도

```python
scores = model.evaluate(X_train, y_train)
print(f"훈련 정확도: {round(scores[1]*100, 2)}", end='\n\n')

scores = model.evaluate(X_test, y_test)
print(f"테스트 정확도: {round(scores[1]*100, 2)}")
```

# Confusion Matrix

![img](https://www.researchgate.net/publication/336402347/figure/fig3/AS:812472659349505@1570719985505/Calculation-of-Precision-Recall-and-Accuracy-in-the-confusion-matrix.ppm)
* 진양성(TP): 실제 클래스가 양성이고 모델도 양성으로 예측
* 위양성(FP): 실제 클래스는 음성이지만 모델이 양성으로 예측
* 진음성(TN): 실제 클래스가 음성이고 모델도 음성으로 예측
* 위음성(FN): 실제 클래스는 양성이지만 모델이 음성으로 예측


정확도(accuracy) : 예측이 정답과 얼마나 정확한가?

정밀도(precision) : 예측한 것중에 정답의 비율은?

재현율(recall) : 찾아야 할 것중에 실제로 찾은 비율은?

F1 Score : 정밀도와 재현율의 평균

* 참고 - https://datascienceschool.net/03%20machine%20learning/09.04%20%EB%B6%84%EB%A5%98%20%EC%84%B1%EB%8A%A5%ED%8F%89%EA%B0%80.html

```python
from sklearn.metrics import confusion_matrix
import sklearn

# y_test_pred = model.predict_classes(X_test)
y_test_pred = model.predict(X_test.values).round()
c_matrix = confusion_matrix(y_test, y_test_pred)

# accurracy, precision, recall, F1-score
print(sklearn.metrics.classification_report(y_test, y_test_pred))
```

```python
# plotly confusion matrix
import plotly.figure_factory as ff

fig = ff.create_annotated_heatmap(c_matrix, 
                                  x=['No Diabetes', 'Diabetes'], 
                                  y=['No Diabetes', 'Diabetes'],
                                  colorscale='blues')

fig.update_layout(title_text='<i><b>Confusion matrix</b></i>',
                  xaxis = dict(title='Prediction'),
                  yaxis = dict(title='Actual'),
                  width=600,
                  height=600)
```
