```python
def candlestick(df, name):
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots

    # Create figure with secondary y-axis
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    # include candlestick with rangeselector
    fig.add_trace(
        go.Candlestick(
            x=df.index,
            open=df['시가'], 
            high=df['고가'],
            low=df['저가'], 
            close=df['종가'],
            name='등락'
        ),
        secondary_y=True
    )

    # include a go.Bar trace for volumes
    fig.add_trace(
        go.Bar(
            x=df.index, 
            y=df['거래량'], 
            name='거래량'),
        secondary_y=False
    )

    fig.layout.yaxis2.showgrid=False
    fig.update_layout(
        title=f'{name} 캔들스틱',
        height=550
    )
    fig.show()
```

```python
candlestick(corps, '현대자동차')
```

# 차동차 관련 업계

```python
vehicle = company_info[company_info['업종'].str.contains('자동차')].reset_index(drop=True)
vehicle.head()
```

```python
vehicle['회사명']
```

```python
car_dic = {}
for corp in vehicle['회사명']:
    car_dic[corp] = get_code(corp)
```

```python
from tqdm.notebook import tqdm
```

```python
corps = pd.DataFrame()

for corp in tqdm(car_dic.keys()):
    corp_code = get_code(corp)
    df = krx.read(corp_code, start=start, end=end)
    df['기업명'] = corp
    df['기업코드'] = corp_code
    corps = corps.append(df)
```
